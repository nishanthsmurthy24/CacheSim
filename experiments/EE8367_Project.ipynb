```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
input_lru = "./experiments_1C/rollup_lru.csv"
output_lru = "./experiments_1C/lru_cleaned.csv"

# Read data from input CSV file
df_lru = pd.read_csv(input_lru)

# Rename the first column based on specified ranges
df_lru.loc[0:139, df_lru.columns[0]] = "SPEC06"
df_lru.loc[140:229, df_lru.columns[0]] = "SPEC17"
df_lru.loc[230:429, df_lru.columns[0]] = "Ligra"
df_lru.loc[430:484, df_lru.columns[0]] = "PARSEC"
df_lru.loc[485:750, df_lru.columns[0]] = "Cloudsuite"

# Write modified data to output CSV file
df_lru.to_csv(output_lru, index=False)

# Read data from input CSV file
df_lru = pd.read_csv(output_lru)

# Filter out rows with Core_0_IPC values equal to 0
df_lru = df_lru[df_lru['Core_0_IPC'] != 0]

# Group the DataFrame by 'Exp' and 'Traces' columns
grouped = df_lru.groupby(['Trace', 'Exp'])

# Calculate speedup relative to the 'nopref' baseline
baseline_lru_ipc = grouped['Core_0_IPC'].mean().loc[:, 'nopref']  # IPC value for 'nopref' experiment in each group
speedup_lru = grouped['Core_0_IPC'].mean() / baseline_lru_ipc
```

```python
srrip_relative = (speedup_srrip / speedup_lru - 1) * 100
srrip_relative = srrip_relative.reset_index()
# Drop rows corresponding to 'nopref' experiment
srrip_relative = srrip_relative[srrip_relative['Exp'] != 'nopref']
# Pivot the DataFrame for better plotting
srrip_relative_pivot = srrip_relative.pivot(index='Trace', columns='Exp', values='Core_0_IPC')

drrip_relative = (speedup_drrip / speedup_lru - 1) * 100
drrip_relative = drrip_relative.reset_index()
# Drop rows corresponding to 'nopref' experiment
drrip_relative = drrip_relative[drrip_relative['Exp'] != 'nopref']
# Pivot the DataFrame for better plotting
drrip_relative_pivot = drrip_relative.pivot(index='Trace', columns='Exp', values='Core_0_IPC')

rlr_relative = (speedup_rlr / speedup_lru - 1) * 100
rlr_relative = rlr_relative.reset_index()
# Drop rows corresponding to 'nopref' experiment
rlr_relative = rlr_relative[rlr_relative['Exp'] != 'nopref']
# Pivot the DataFrame for better plotting
rlr_relative_pivot = rlr_relative.pivot(index='Trace', columns='Exp', values='Core_0_IPC')

ship_relative = (speedup_ship / speedup_lru - 1) * 100
ship_relative = ship_relative.reset_index()
# Drop rows corresponding to 'nopref' experiment
ship_relative = ship_relative[ship_relative['Exp'] != 'nopref']
# Pivot the DataFrame for better plotting
ship_relative_pivot = ship_relative.pivot(index='Trace', columns='Exp', values='Core_0_IPC')

ship_pp_relative = (speedup_ship_pp / speedup_lru - 1) * 100
ship_pp_relative = ship_pp_relative.reset_index()
# Drop rows corresponding to 'nopref' experiment
ship_pp_relative = ship_pp_relative[ship_pp_relative['Exp'] != 'nopref']
# Pivot the DataFrame for better plotting
ship_pp_relative_pivot = ship_pp_relative.pivot(index='Trace', columns='Exp', values='Core_0_IPC')

hawkeye_relative = (speedup_rlr / speedup_lru - 1) * 100
hawkeye_relative = hawkeye_relative.reset_index()
# Drop rows corresponding to 'nopref' experiment
hawkeye_relative = hawkeye_relative[hawkeye_relative['Exp'] != 'nopref']
# Pivot the DataFrame for better plotting
hawkeye_relative_pivot = hawkeye_relative.pivot(index='Trace', columns='Exp', values='Core_0_IPC')
```

```python
fig, axs = plt.subplots(2, 3, figsize=(12, 6))  # Creating subplots with 2 rows and 1 column

# Plotting for SR-RIP
srrip_relative_pivot.plot(kind='bar', ax=axs[0][0], color=['lightgrey', 'darkgrey', 'black', 'grey'])
axs[0][0].set_title('Relative Speedup - SR-RIP')
axs[0][0].set_ylabel('Speedup')
axs[0][0].set_xticklabels(srrip_relative_pivot.index, rotation=45, ha='right')
axs[0][0].grid(axis='y', linestyle='-.')
axs[0][0].set_ylim(-5.0, 1.0)

# Move legend outside
axs[0][0].legend().set_visible(False)

# Plotting for DR-RIP
drrip_relative_pivot.plot(kind='bar', ax=axs[0][1], color=['lightgrey', 'darkgrey', 'black', 'grey'])
axs[0][1].set_title('Relative Speedup - DR-RIP')
axs[0][1].set_xticklabels(drrip_relative_pivot.index, rotation=45, ha='right')
axs[0][1].grid(axis='y', linestyle='-.')
axs[0][1].set_ylim(-5.0, 1.0)

# Move legend outside
axs[0][1].legend().set_visible(False)

# Plotting for RLR
rlr_relative_pivot.plot(kind='bar', ax=axs[0][2], color=['lightgrey', 'darkgrey', 'black', 'grey'])
axs[0][2].set_title('Relative Speedup - RLR')
axs[0][2].set_xticklabels(rlr_relative_pivot.index, rotation=45, ha='right')
axs[0][2].grid(axis='y', linestyle='-.')
axs[0][2].set_ylim(-5.0, 1.0)

# Move legend outside
axs[0][2].legend(title='Experiment', bbox_to_anchor=(1.05, 1), loc='upper left')

# Plotting for SHiP
ship_relative_pivot.plot(kind='bar', ax=axs[1][0], color=['lightgrey', 'darkgrey', 'black', 'grey'])
axs[1][0].set_title('Relative Speedup - SHiP')
axs[0][0].set_ylabel('Speedup')
axs[1][0].set_xticklabels(ship_relative_pivot.index, rotation=45, ha='right')
axs[1][0].grid(axis='y', linestyle='-.')
axs[1][0].set_ylim(-5.0, 1.0)

# Move legend outside
axs[1][0].legend().set_visible(False)

# Plotting for SHiP++
ship_pp_relative_pivot.plot(kind='bar', ax=axs[1][1], color=['lightgrey', 'darkgrey', 'black', 'grey'])
axs[1][1].set_title('Relative Speedup - SHiP++')
axs[1][1].set_xticklabels(ship_pp_relative_pivot.index, rotation=45, ha='right')
axs[1][1].grid(axis='y', linestyle='-.')
axs[1][1].set_ylim(-5.0, 1.0)

# Move legend outside
axs[1][1].legend().set_visible(False)

# Plotting for Hawkeye
hawkeye_relative_pivot.plot(kind='bar', ax=axs[1][2], color=['lightgrey', 'darkgrey', 'black', 'grey'])
axs[1][2].set_title('Relative Speedup - Hawkeye')
axs[1][2].set_xticklabels(hawkeye_relative_pivot.index, rotation=45, ha='right')
axs[1][2].grid(axis='y', linestyle='-.')
axs[1][2].set_ylim(-5.0, 1.0)

axs[1][2].legend(title='Experiment', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
